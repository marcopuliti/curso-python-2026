## 1. ¿Por qué funciones?

La semana pasada convertimos Celsius a Fahrenheit. Si necesitáramos hacerlo en 5 lugares distintos del programa, ¿copiamos la fórmula 5 veces? No: la **encapsulamos en una función** y la llamamos por su nombre.

Una función es una **receta con nombre**: recibe ingredientes (*parámetros*), hace algo, y (generalmente) devuelve un resultado (*retorno*).

```python
def nombre_de_la_funcion(parametro1, parametro2):
    # bloque indentado: el cuerpo de la función
    resultado = ...
    return resultado
```

Ya venís usando funciones que escribió otra gente: `print()`, `type()`, `input()`, `range()`. Hoy aprendés a escribir las tuyas.

````python
def celsius_a_fahrenheit(celsius):
    fahrenheit = 9 / 5 * celsius + 32
    return fahrenheit

# Definirla no la ejecuta. Hay que LLAMARLA:
print(celsius_a_fahrenheit(0))
print(celsius_a_fahrenheit(100))
print(celsius_a_fahrenheit(36.5))
````

Observá la anatomía:

- `def` anuncia que estamos *definiendo* una función.
- `celsius` es el **parámetro**: una variable que toma el valor que le pasemos al llamar.
- `return` indica qué valor devuelve la función. Al ejecutarse el `return`, la función termina.
- El valor devuelto se puede guardar en una variable: `f = celsius_a_fahrenheit(25)`.

### `return` vs. `print`

Error clásico de principiante: confundir *devolver* con *mostrar*.

- `print` solo **muestra** en pantalla; el valor no se puede reutilizar.
- `return` **devuelve** el valor al que llamó la función, para seguir calculando con él.

Regla práctica: **las funciones calculan y devuelven; el `print` va afuera.**

````python
def cuadrado_que_devuelve(x):
    return x ** 2

def cuadrado_que_imprime(x):
    print(x ** 2)

# Con return puedo encadenar cálculos:
print(cuadrado_que_devuelve(3) + cuadrado_que_devuelve(4))   # 9 + 16 = 25

# Con print no: la función devuelve None ('nada')
resultado = cuadrado_que_imprime(3)
print(resultado)
````

## 2. Parámetros con valor por defecto y argumentos con nombre

Un parámetro puede tener un **valor por defecto**, que se usa si quien llama no lo especifica. Y al llamar, podemos nombrar los argumentos para más claridad (y para pasarlos en cualquier orden):

````python
def caida_libre(tiempo, gravedad=9.81):
    "Distancia recorrida en caída libre, en metros."
    return 0.5 * gravedad * tiempo ** 2

print(caida_libre(2))                  # usa g de la Tierra
print(caida_libre(2, gravedad=1.62))   # en la Luna
print(caida_libre(gravedad=3.71, tiempo=2))  # en Marte, argumentos con nombre
````

## 3. Documentar con docstrings

Un **docstring** es un texto entre triples comillas en la primera línea de la función. Describe qué hace, qué recibe y qué devuelve. Es lo que aparece cuando alguien pide ayuda con `help()`:

````python
def indice_masa_corporal(peso, altura):
    """Calcula el índice de masa corporal (IMC).

    Parámetros:
        peso: masa corporal en kilogramos.
        altura: estatura en metros.

    Devuelve:
        El IMC como float (kg/m2).
    """
    return peso / altura ** 2

help(indice_masa_corporal)
print(f'IMC: {indice_masa_corporal(70, 1.75):.1f}')
````

## 4. Ámbito de variables: locales y globales

Las variables creadas **dentro** de una función son **locales**: nacen al llamar la función y mueren cuando termina. No existen afuera.

````python
def duplicar(x):
    doble = 2 * x   # 'doble' y 'x' son locales
    return doble

print(duplicar(21))

# Si descomentás la línea siguiente, da NameError: 'doble' no existe acá afuera
# print(doble)
````

````python
# Las variables de afuera (globales) se pueden LEER dentro de una función...
gravedad = 9.81

def peso(masa):
    return masa * gravedad   # lee la global

print(peso(70))

# ...pero modificar globales dentro de funciones es mala práctica:
# hace el código difícil de seguir. Preferí siempre parámetros y return.
````

## 5. Descomponer problemas: funciones que llaman funciones

El verdadero poder: resolver un problema grande como composición de problemas chicos. Cada función hace **una sola cosa** y la hace bien.

````python
def es_primo(n):
    """Devuelve True si n es un número primo."""
    if n < 2:
        return False
    for divisor in range(2, n):
        if n % divisor == 0:
            return False   # encontramos un divisor: no es primo
    return True            # sobrevivió a todos los divisores

def contar_primos(limite):
    """Cuenta los primos entre 2 y limite (inclusive)."""
    contador = 0
    for n in range(2, limite + 1):
        if es_primo(n):
            contador += 1
    return contador

print(es_primo(7), es_primo(10), es_primo(97))
print(f'Hay {contar_primos(100)} primos hasta 100')   # debe dar 25
````

## 6. Del notebook al script `.py`

Los notebooks son ideales para explorar, pero el software científico se distribuye en **scripts**: archivos de texto con extensión `.py` que se ejecutan completos, de arriba hacia abajo.

3. Ejecutalo: botón ▶ de VS Code, o en la terminal escribí `python tablas.py`.

### La diferencia clave

| Notebook (`.ipynb`) | Script (`.py`) |
|---|---|
| Celdas que se ejecutan a demanda, en cualquier orden | Se ejecuta completo, de arriba hacia abajo |
| Ideal para explorar, probar, graficar | Ideal para automatizar y compartir |
| El estado puede quedar 'sucio' entre celdas | Cada ejecución arranca de cero |
