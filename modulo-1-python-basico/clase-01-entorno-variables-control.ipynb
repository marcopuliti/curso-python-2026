## 2. Variables: ponerle nombre a los datos

Una **variable** es un nombre que apunta a un valor. Se crea con el signo `=` (que acá significa *asignación*, no igualdad matemática).

```python
temperatura = 23.5        # grados Celsius
muestras = 120            # cantidad de mediciones
laboratorio = 'Bloque II' # texto
calibrado = True          # verdadero o falso

print(temperatura)
print(muestras)
print(laboratorio)
print(calibrado)
```

Reglas para los nombres de variables:

- Pueden tener letras, números y guiones bajos, pero **no pueden empezar con número**: `dato_1` ✅, `1_dato` ❌.
- No pueden tener espacios ni tildes (se recomienda evitar la ñ).
- Python distingue mayúsculas: `Temperatura` y `temperatura` son variables distintas.
- Usá nombres descriptivos: `velocidad_inicial` dice mucho más que `v0` o `x`.

### Tipos de datos básicos

Cada valor tiene un **tipo**. Los cuatro fundamentales son:

| Tipo | Nombre en Python | Ejemplos |
|---|---|---|
| Entero | `int` | `120`, `-3`, `0` |
| Decimal (punto flotante) | `float` | `23.5`, `-0.001`, `6.022e23` |
| Texto (cadena) | `str` | `'Bloque II'`, `"agua"` |
| Lógico (booleano) | `bool` | `True`, `False` |

La función `type()` nos dice el tipo de cualquier cosa:

```python
print(type(temperatura))
print(type(muestras))
print(type(laboratorio))
print(type(calibrado))

# Notación científica: el número de Avogadro
avogadro = 6.022e23
print(avogadro, type(avogadro))
```

### Conversiones entre tipos

A veces necesitamos convertir un tipo en otro. Las funciones `int()`, `float()` y `str()` hacen exactamente eso:

```python
edad_texto = '34'           # esto es texto, no un número
edad = int(edad_texto)      # ahora sí es un entero
print(edad + 1)

pi_aproximado = float('3.14')
print(pi_aproximado * 2)

# Ojo: esto daría error si lo descomentás, porque 'hola' no es un número
# numero = int('hola')
```

## 3. Operadores: hacer cuentas

### Operadores aritméticos

| Operador | Operación | Ejemplo | Resultado |
|---|---|---|---|
| `+` | suma | `7 + 2` | `9` |
| `-` | resta | `7 - 2` | `5` |
| `*` | multiplicación | `7 * 2` | `14` |
| `/` | división | `7 / 2` | `3.5` |
| `//` | división entera | `7 // 2` | `3` |
| `%` | resto (módulo) | `7 % 2` | `1` |
| `**` | potencia | `7 ** 2` | `49` |

Dos de estos van a resultar nuevos incluso si usás calculadora todos los días: `//` y `%`. El resto de la división (`%`) es utilísimo: por ejemplo, `n % 2 == 0` dice si `n` es par.

```python
# Energía cinética: E = (1/2) m v**2
masa = 2.0        # kg
velocidad = 3.5   # m/s
energia = 0.5 * masa * velocidad ** 2
print(energia)    # en Joules

# División entera y resto: ¿cuántas horas y minutos son 200 minutos?
minutos_totales = 200
horas = minutos_totales // 60
minutos = minutos_totales % 60
print(horas, 'horas y', minutos, 'minutos')
```

### f-strings: mostrar resultados con formato

Para armar mensajes que mezclan texto y variables, lo más cómodo son las **f-strings**: un string que empieza con `f` y permite poner variables entre llaves `{}`.

```python
print(f'La energía cinética es {energia} J')

# Con formato: :.2f significa 'mostrar 2 decimales'
pi = 3.14159265358979
print(f'Pi con dos decimales: {pi:.2f}')
print(f'Pi con cinco decimales: {pi:.5f}')
print(f'Avogadro en notación científica: {6.022e23:.3e}')
```

### Operadores de comparación y lógicos

Las comparaciones producen un resultado booleano (`True` o `False`):

| Operador | Significado |
|---|---|
| `==` | igual a (¡dos signos! uno solo es asignación) |
| `!=` | distinto de |
| `<`, `>` | menor, mayor |
| `<=`, `>=` | menor o igual, mayor o igual |

Y se combinan con los operadores lógicos `and`, `or` y `not`:

```python
ph = 6.8

print(ph == 7)              # ¿es exactamente 7?
print(ph < 7)               # ¿es ácido?
print(6.5 <= ph <= 7.5)     # ¿está en el rango aceptable? (¡se puede encadenar!)
print(ph < 7 and ph > 6)    # combinación con and
print(not ph > 7)           # negación
```

## 4. Condicionales: tomar decisiones

Con `if` el programa ejecuta un bloque de código **solo si** se cumple una condición. La estructura completa es `if` / `elif` ("else if", opcional, las que hagan falta) / `else` (opcional).

⚠️ **La indentación (sangría) es obligatoria en Python**: las líneas que están "adentro" del `if` van desplazadas 4 espacios. Así Python sabe dónde empieza y termina cada bloque.

```python
ph = 6.8

if ph < 7:
    print('La solución es ácida')
elif ph == 7:
    print('La solución es neutra')
else:
    print('La solución es alcalina')
```

```python
# Las condiciones se pueden combinar
magnitud = 5.8
profundidad = 12  # km

if magnitud >= 5.5 and profundidad < 70:
    print('Sismo fuerte y superficial: revisar protocolos')
else:
    print('Sin alerta especial')
```

## 5. Bucles: repetir sin copiar y pegar

Las computadoras brillan repitiendo tareas. Python tiene dos bucles:

### `while`: repetir mientras se cumpla una condición

```python
# Decaimiento: ¿cuántos períodos tarda una muestra en caer por debajo del 1%?
cantidad = 100.0   # porcentaje inicial
periodos = 0

while cantidad >= 1.0:
    cantidad = cantidad / 2   # en cada período se desintegra la mitad
    periodos = periodos + 1

print(f'Hacen falta {periodos} períodos (queda {cantidad:.3f}%)')
```

⚠️ Cuidado con los **bucles infinitos**: si la condición nunca deja de cumplirse, el programa no termina nunca. Si te pasa, frenalo con el botón ⏹ (interrupt) del notebook.

### `for` + `range`: repetir un número conocido de veces

`range(n)` genera los enteros `0, 1, 2, ..., n-1`. También acepta inicio y paso: `range(inicio, fin, paso)`.

```python
for i in range(5):
    print(f'Medición número {i}')

print('---')

# range con inicio y paso: los pares del 2 al 10
for n in range(2, 11, 2):
    print(n)
```

```python
# El patrón más usado de la ciencia computacional: el ACUMULADOR
# Sumar 1 + 2 + ... + 100 (como Gauss, pero sin la viveza)
suma = 0
for n in range(1, 101):
    suma = suma + n   # también se escribe: suma += n

print(f'La suma de 1 a 100 es {suma}')
```

### `break` y `continue`

- `break` corta el bucle inmediatamente.
- `continue` salta a la siguiente vuelta.

Se usan poco, pero conviene reconocerlos:

```python
# Buscar el primer múltiplo de 7 mayor que 100
for n in range(101, 200):
    if n % 7 == 0:
        print(f'Encontrado: {n}')
        break
```
