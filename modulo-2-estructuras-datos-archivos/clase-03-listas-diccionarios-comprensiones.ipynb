## 1. Listas: muchos valores en una sola variable

Una **lista** es una secuencia ordenada y modificable de valores, entre corchetes:

````python
temperaturas = [22.4, 21.8, 23.1, 24.0, 22.9]

print(temperaturas)
print(len(temperaturas))      # cantidad de elementos
print(temperaturas[0])        # primer elemento: ¡se cuenta desde 0!
print(temperaturas[4])        # último (índice 4 en una lista de 5)
print(temperaturas[-1])       # también el último: índices negativos cuentan desde el final
````

### Rebanadas (slicing)

`lista[desde:hasta]` extrae una sublista, **incluyendo** `desde` y **excluyendo** `hasta` (igual que `range`):

````python
print(temperaturas[1:3])    # índices 1 y 2
print(temperaturas[:2])     # los dos primeros
print(temperaturas[2:])     # del índice 2 en adelante
print(temperaturas[::2])    # uno de cada dos (el tercer número es el paso)
````

### Modificar listas

Las listas son **mutables**: se pueden cambiar después de creadas.

````python
mediciones = [3.2, 3.5, 3.1]

mediciones.append(3.8)        # agregar al final
print(mediciones)

mediciones[0] = 3.25          # reemplazar por índice
print(mediciones)

mediciones.remove(3.5)        # quitar la primera aparición de un valor
print(mediciones)

ultimo = mediciones.pop()     # quitar y devolver el último
print(ultimo, mediciones)

mediciones.extend([4.0, 4.1]) # agregar varios
print(mediciones)
````

### Funciones útiles sobre listas

`len`, `sum`, `min`, `max` y `sorted` hacen el 80 % del trabajo estadístico básico:

````python
temperaturas = [22.4, 21.8, 23.1, 24.0, 22.9]

print(f'n = {len(temperaturas)}')
print(f'media = {sum(temperaturas) / len(temperaturas):.2f}')
print(f'mínimo = {min(temperaturas)}, máximo = {max(temperaturas)}')
print(f'ordenadas: {sorted(temperaturas)}')
print(f'la original no cambió: {temperaturas}')

# temperaturas.sort() en cambio SÍ modifica la lista original
temperaturas.sort()
print(f'ahora sí: {temperaturas}')
````

### Recorrer listas

El `for` de Python recorre directamente los elementos (no hacen falta índices). Y cuando necesitás el índice además del valor, está `enumerate`:

````python
for t in temperaturas:
    print(f'{t} °C → {9 / 5 * t + 32:.1f} °F')

print('---')

for i, t in enumerate(temperaturas):
    print(f'Medición {i}: {t} °C')
````

## 2. Tuplas: secuencias que no cambian

Una **tupla** es como una lista pero **inmutable** (no se puede modificar). Se escribe con paréntesis. Se usa para datos que van juntos y no deberían cambiar: coordenadas, pares (valor, error), fechas.

````python
punto = (32.5, -65.8)            # latitud, longitud
medicion = ('Sitio A', 22.4, 0.1) # sitio, valor, error

# Desempaquetado: asignar los elementos a variables de una vez
lat, lon = punto
sitio, valor, error = medicion
print(f'{sitio}: {valor} ± {error} (en {lat}, {lon})')

# Si descomentás esto da TypeError: las tuplas no se modifican
# punto[0] = 33.0
````

El desempaquetado ya lo venías usando sin saberlo: `for i, t in enumerate(...)` desempaqueta tuplas `(índice, valor)`. Y una lista de tuplas es una mini-tabla de datos:

````python
muestras = [
    ('S-01', 6.8),
    ('S-02', 7.4),
    ('S-03', 5.9),
    ('S-04', 7.1),
]

for codigo, ph in muestras:
    print(f'{codigo}: pH {ph}')
````

## 3. Diccionarios: datos con nombre

Un **diccionario** asocia **claves** con **valores**: como un fichero donde cada ficha tiene una etiqueta. Se escribe con llaves `{clave: valor}`. Es LA estructura para datos heterogéneos:

````python
muestra = {
    'codigo': 'S-01',
    'ph': 6.8,
    'temperatura': 22.4,
    'sitio': 'Arroyo El Volcán',
}

print(muestra['ph'])           # acceder por clave

muestra['profundidad'] = 1.5   # agregar una clave nueva
muestra['ph'] = 6.9            # modificar un valor
print(muestra)
````

````python
# Acceder a una clave que no existe da KeyError. Para eso está .get():
print(muestra.get('conductividad'))          # None si no existe
print(muestra.get('conductividad', 0.0))     # o un valor por defecto

# Preguntar si una clave existe:
print('ph' in muestra)
print('conductividad' in muestra)
````

### Recorrer diccionarios

````python
for clave in muestra:                  # recorre las claves
    print(clave)

print('---')

for clave, valor in muestra.items():   # claves y valores a la vez
    print(f'{clave}: {valor}')
````

### El patrón CONTADOR

Tan importante como el acumulador de la clase 1. ¿Cuántas veces aparece cada cosa? Un diccionario donde la clave es la cosa y el valor es el conteo:

````python
avistajes = ['cóndor', 'puma', 'cóndor', 'zorro', 'cóndor', 'puma', 'guanaco']

conteo = {}
for especie in avistajes:
    conteo[especie] = conteo.get(especie, 0) + 1

print(conteo)

# ¿Cuál fue la más vista? max con key= busca el máximo según el conteo
mas_vista = max(conteo, key=conteo.get)
print(f'La más avistada: {mas_vista} ({conteo[mas_vista]} veces)')
````

## 4. Conjuntos: elementos únicos

Un **conjunto** (`set`) guarda elementos **sin repetidos y sin orden**. Sirve para eliminar duplicados y para las operaciones de la teoría de conjuntos que ya conocés: unión, intersección, diferencia.

````python
# Especies registradas en dos sitios de muestreo
sitio_a = {'cóndor', 'puma', 'zorro', 'guanaco'}
sitio_b = {'puma', 'zorro', 'liebre', 'jote'}

print(sitio_a & sitio_b)   # intersección: en ambos sitios
print(sitio_a | sitio_b)   # unión: en alguno de los dos
print(sitio_a - sitio_b)   # diferencia: solo en A
print(sitio_a ^ sitio_b)   # diferencia simétrica: en uno solo de los dos
````

````python
# set() también elimina duplicados de una lista:
lecturas = [1, 2, 2, 3, 1, 4, 2, 5, 3]
valores_distintos = set(lecturas)
print(valores_distintos)
print(f'Hubo {len(valores_distintos)} valores distintos en {len(lecturas)} lecturas')
````

## 5. Comprensiones de listas: transformar y filtrar en una línea

El patrón "crear lista vacía + for + append" es tan común que Python tiene una sintaxis compacta para escribirlo:

```python
# Versión larga                        # Versión con comprensión
resultado = []
for x in datos:                        resultado = [x * 2 for x in datos]
    resultado.append(x * 2)
```

Se lee casi como matemática: $\{2x : x \in datos\}$.

````python
celsius = [22.4, 21.8, 23.1, 24.0, 22.9]

# Transformar: todas a Fahrenheit
fahrenheit = [9 / 5 * t + 32 for t in celsius]
print(fahrenheit)

# Filtrar: solo las mayores a 22.5 (se agrega un if al final)
calidas = [t for t in celsius if t > 22.5]
print(calidas)

# Transformar Y filtrar a la vez
calidas_f = [9 / 5 * t + 32 for t in celsius if t > 22.5]
print(calidas_f)
````

````python
# Funcionan de maravilla con listas de tuplas:
muestras = [('S-01', 6.8), ('S-02', 7.4), ('S-03', 5.9), ('S-04', 7.1)]

acidas = [codigo for codigo, ph in muestras if ph < 7]
print(f'Muestras ácidas: {acidas}')

# También hay comprensiones de diccionarios:
ph_por_codigo = {codigo: ph for codigo, ph in muestras}
print(ph_por_codigo)
````

### Generadores (para saber que existen)

Si cambiás los corchetes por paréntesis, obtenés un **generador**: produce los valores de a uno, sin armar toda la lista en memoria. Con millones de datos, la diferencia importa. Por ahora alcanza con reconocerlos:

````python
# suma de los cuadrados del 1 al millón, sin crear una lista de un millón de elementos
total = sum(x ** 2 for x in range(1, 1_000_001))
print(total)
````
